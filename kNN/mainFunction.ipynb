### 数据可视化

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
#按不同标签以不同颜色显示
exampleLabel = np.array(exampleLabel)
label1 = np.where(exampleLabel.ravel() == 1)
ax.scatter(exampleMat[label1,0],exampleMat[label1,1],marker='x',color = 'r',label = 'didnt like=1')
label2 = np.where(exampleLabel.ravel() == 2)
ax.scatter(exampleMat[label2,0],exampleMat[label2,1],marker='o',color = 'b',label = 'smallDoses=2')
label3 = np.where(exampleLabel.ravel() == 3)
ax.scatter(exampleMat[label3,0],exampleMat[label3,1],marker='*',color = 'g',label = 'largeDoses=3')
ax.set_title('data visualization')
ax.set_xlabel('Mileage of flight')
ax.set_ylabel('video game percent\'%\'')
plt.legend(loc = 'upper left')
plt.show()
```
